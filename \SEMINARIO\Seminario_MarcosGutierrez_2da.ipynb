```
numeros = [str(i) for i in range(1,10)]
operadores = ["+","-","/","*"]
print(numeros)
print(operadores)
```

Aplica el algoritmo al juego de datos generado

```
#libreria necesaria
from itertools import permutations
```

Respuesta

```
def operaciones(numeros, operadores, objetivo):
  res = []
  for val in permutations(numeros,5):
    for oper in permutations(operadores,4):
        formula = "".join(oper+val for oper,val in zip([""]+list(oper),val))
        if eval(formula) == objetivo: res.append(formula)
  return(set(res))
```

```
operaciones(numeros,operadores,4) #Esta funcion brinda todas las operaciones que dan valores iguales a un objetivo especifico. 
```

```
#Maximos y minimos enteros:
def maximo(numeros,operadores):
  res = [9]
  for val in permutations(numeros,5):
    for oper in permutations(operadores,4):
        formula = "".join(oper+val for oper,val in zip([""]+list(oper),val))
        if eval(formula) > res[-1] and eval(formula).is_integer == True: res.append(eval(formula)) #Si se desea ver el valor maximo real remover la porcion de codigo and eval(formula).is_integer == True
  return(res[-1])


def minimo(numeros,operadores):
  res = [1]
  for val in permutations(numeros,5):
    for oper in permutations(operadores,4):
        formula = "".join(oper+val for oper,val in zip([""]+list(oper),val))
        if eval(formula) < res[-1] and eval(formula).is_integer == True: res.append(eval(formula)) #Si se desea ver el valor minimo real remover la porcion de codigo and eval(formula).is_integer == True
  return(res[-1]) 
```

```
#Valor entero maximo
maximo(numeros,operadores)
```

```
#Valor entero minimo
minimo(numeros,operadores)
```
